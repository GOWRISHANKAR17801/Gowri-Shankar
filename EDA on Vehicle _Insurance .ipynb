```python
def finding_outliers(df):
    q1 = df.quantile(0.25)
    q3 = df.quantile(0.75)
    iqr = q3-q1
    outlier=df[((df<q1-1.5*iqr))|(df>(q3+1.5*iqr))]
    return outlier
```

```python
outliers_Age = finding_outliers(df1['Age'])
print(len(outliers_Age))
print(outliers_Age.max())
print(outliers_Age.min())
```

```python
outliers_regioncode = finding_outliers(df1['Region Code'])
print(len(outliers_regioncode))
print(outliers_regioncode.max())
print(outliers_regioncode.min())
```

```python
outliers_Annualpremium = finding_outliers(df2["annual_premium(inRs)"])
print("Length : ",len(outliers_Annualpremium))
print("Maximum :",outliers_Annualpremium.max())
print("Minimum :",outliers_Annualpremium.min())
```

```python
# Solving the outliers
Q1 = df2["annual_premium(inRs)"].quantile(0.25)
Q3 = df2["annual_premium(inRs)"].quantile(0.75)
IQR = Q3-Q1
lower_bound = (Q1 - (1.5*IQR))
upper_bound = (Q3 + (1.5*IQR))
#Identify the outliers
outliers = df2[(df2['annual_premium(inRs)'] < lower_bound) | (df2["annual_premium(inRs)"]> upper_bound)]
```
